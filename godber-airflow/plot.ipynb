```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('temps.fwf', delim_whitespace=True, usecols=[2, 17, 18], names=['date', 'max_temp', 'min_temp'])

df2 = df.set_index(pd.to_datetime(df.date, format='%Y%m%d'))

df3 = df2.drop(['date'], axis=1)

ax = df3.plot(figsize=(11, 8.5))
fig = ax.get_figure()
fig.savefig('out.png')
```
